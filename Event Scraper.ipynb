```python
import numpy as np
import time
import csv
import pandas as pd
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import NavigableString
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup as bs
from selenium.webdriver.support.wait import WebDriverWait
options = Options()
driver=webdriver.Chrome(ChromeDriverManager().install(),options=options)

driver.get('https://www.worldconferencealerts.com/event.php')
driver.maximize_window()
time.sleep(2)
Event_name=[]
Date_and_Time=[]
Event_Description=[]
Event_Location=[]

for path in ["//*[@id='searchmonth']/option[7]","//*[@id='searchmonth']/option[8]"]:
    
    driver.find_element_by_xpath('//*[@id="searchcountry_chosen"]/a').click()
    time.sleep(2)
    driver.find_element_by_xpath('//*[@id="searchcountry_chosen"]/div/ul/li[99]').click()
    time.sleep(2)

    driver.find_element_by_xpath('//*[@id="searchtopic_chosen"]').click()
    time.sleep(2)
    driver.find_element_by_xpath('//*[@id="searchtopic_chosen"]/div/ul/li[53]').click()
    time.sleep(2)
    driver.find_element_by_xpath('//*[@id="searchmonth"]').click()
    time.sleep(2)
    driver.find_element_by_xpath(path).click()
    time.sleep(2)
    driver.find_element_by_xpath('//*[@id="advance_search"]/div[2]/div[4]/button').click()
    time.sleep(2)
    for i in range(0,50):
        
        try:
            
            driver.find_element_by_xpath('//*[@id="loadmorebtn"]').click()
   
        except:
            break
    time.sleep(2)
    confe_page=driver.page_source
    conference_html = bs(confe_page, "html.parser")
    confe_1=conference_html.findAll("div", {"class":"conf-list-detal"})
    
    

    
    for i in confe_1:
        try:
            Event_name.append(i.find_all('a',{'class':'conf-evnt-title'})[0].text.strip())
        except:
            Event_name.append(np.nan)

        try:
            Date_and_Time.append(i.find_all("span",{"itemprop":"startDate"})[0].text.strip().replace("\n"," "))
        except:
            Date_and_Time.append(np.nan)

        try:
            Event_Description.append(i.find_all('p',{'class':'conf-evnt-pargha'})[0].text.strip())
        except:
            Event_Description.append(np.nan)

        try:
            Event_Location.append(i.find_all("span",{"itemprop":"areaServed"})[0].text)
        except:
            Event_Location.append(np.nan)
    time.sleep(10)
     
df1=pd.DataFrame({' Event_name': Event_name,
        'Date_and_Time':Date_and_Time,
        'Event_Description':Event_Description,
        'Event_Location':Event_Location
        })

df1.head()
driver.close()
```
